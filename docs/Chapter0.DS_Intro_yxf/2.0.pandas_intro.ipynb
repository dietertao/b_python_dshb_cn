## Series（智能货架）

`Series` 用来装一维数据（a single dimension of data）。
- 可以想象表格中的一列，或Python的list列表数据。（提问：Python， Numpy有一堆可以存储一维数据的结构，为什么还需要Pandas Series？）
- Pandas赋予了Series一些特殊的关键特征


特征：
- 带标签的一维数据容器
- 每个「货架格子」有唯一标签（索引）
- 支持混合类型（但推荐同类型数据）

 


### 举例说明Serie

````python
# 举例：创建一个简单的Series，存储温度数据 
my_series = pd.Series([97.5, 98.2, 99.5, 98.1, 98.7], name='temperatures')

# print the series 
my_series
````

代码解释：调用Pandas中的 `Series()` construct, 输入一个Python list= 浮点数类型的数值列表给创建方法， 将浮动数据付名为"temperatures." 打印数据时发现不光是一列数值，而是2列数值，还有数据命名和相关的数据类型dtype:`float64`. 
这个例子表明，创建Series，我们可以在存储数据的基础上再附带一点有用的“私货”： **index**/索引/标签/metadata/. 目的：允许人们使用一个identifier（且是显性的）来访问一个数据元素。 


虽然我们可以直接使用Python lists类似的隐形索引来访问`Series`元素，比如： 

````python
my_series[3]
````

也许你会有这个疑问：既然可以直接用Python隐性Index来访问，为何Pandas要特别创建一个单独的explicit index？这就是Pandas看数据的底层思维的不同（see and do things differently). 因为它可以灵活的使用字符串（strings）,时间/日期 （dates）, 甚至任意排序/无序的类型来定义索引=也就是认为地给你的数据匹配一个你认为最合适的说明或标签. 

````python
# 举例： 用Series来创建水果价格货架  
price_series = pd.Series(  
    data=[3.5, 2.8, 4.1],  
    index=["苹果", "香蕉", "橙子"],  
    name="价格"  
) 

price_series
````

````python
# 使用标签来直接访问Series元素
price_series["香蕉"]
````

## DataFrames智能仓库）

特征：
- 多个Series组成的二维表格。（Series可看成是表格中一列数据，DataFrame就是多列数据组成的表格数据结构）
- 行索引（货架位置） + 列索引（商品分类）
  - rows/column index又分别叫做axis 0， axis 1
- 每列可是独立数据类型（数字/文字/日期共存）


### **dataframe** 举例

````python
# 创建水果仓库， 导入的数据是Python字典类型
fruit_warehouse = pd.DataFrame({  
    "价格": [3.5, 2.8, 4.1],  
    "库存量": [120, 90, 75],  
    "产地": ["山东", "海南", "江西"]  
}, index=["货架A", "货架B", "货架C"])  

fruit_warehouse
````

## 四大操作原则（仓库管理手册）
- **原则1**：标签优先访问
```python
# 按货架标签取整行  
fruit_warehouse.loc["货架B"]  

# 按列标签+条件筛选  
fruit_warehouse[fruit_warehouse["价格"] > 3]  
```


- **原则2**：智能处理缺失值
```python  
# 检测空货架格子  
fruit_warehouse.isnull()  

# 填充策略  
fruit_warehouse.fillna({"库存量": 0})  # 库存空缺填0  
fruit_warehouse.dropna(axis=1)       # 删除整列空货架  
```

- **原则3**：轴方向明确操作
  - 密诀：
    - axis=0 → 垂直方向（操作行）
    - axis=1 → 水平方向（操作列）
```python
# 每列平均（沿垂直方向压缩）  
fruit_warehouse.mean(axis=0)  

# 删除库存量列（水平方向移除）  
fruit_warehouse.drop("库存量", axis=1)  
```

- **原则4**：链式方法组合
```python
# 数据清洗流水线  
(fruit_warehouse  
 .query("价格 > 3")             # 筛选高价水果  
 .assign(总价值=lambda df: df["价格"] * df["库存量"])  # 新增列  
 .groupby("产地")["总价值"].sum()  # 按产地汇总  
)  
```

# 与NumPy协作策略

## 转换场景与技巧

| 场景	|  Pandas → NumPy	|  NumPy → Pandas |
| --- | --- | ---  |
| 数学计算 |	df.values 提取数组 |	pd.DataFrame(arr) 基础转换 |
| 保留元数据 |	df.to_numpy() |	添加columns和index参数 |
| 性能优化 |	类型一致性检查后再转换 | 	预处理后再封装| 

典型工作流：
```python
# Pandas清洗 → NumPy计算 → Pandas展示  
clean_data = sales.dropna().astype(np.float32)  
matrix = clean_data.values  
result = np.linalg.eig(matrix)  # NumPy计算特征值  
pd.DataFrame(result, index=clean_data.index)  # 结果转回DataFrame  
```
