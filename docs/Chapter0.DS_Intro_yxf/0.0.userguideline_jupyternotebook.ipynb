## 0. 运行环境设置 Set Up Environment <a id="0"></a>

### 安装Anaconda，并从conda终端安装Jupyter Notebook
- 开启conda终端窗口
  - Windows Exploerer> 右键> Open Conda Prompt
- 如有必要，设置自己的Python虚拟开发环境。Anaconda设置的默认环境为base
  - 输入指令(自设参数： 环境名，python版本)： conda create --name mypy python==3.11
  - 激活虚拟环境：  conda activate mypy
- 启动（网络版）Notebook： 在终端窗口转换到目标目录下，输入指令： jupyter notebook

### VS Code中运行Jupyter Notebook
- 安装VS Code 
- Extensions> 增添Python Extensions for VS Code, Python Debugger Extensions for VS Code
- 设置Python Interpreter
- 新建或打开已存的notebook文件（后缀 filename.ipynb）

## 1. 基本规则 Basic Rules  <a id="1"></a>

### 用好学习笔记，记录使用流程： 享受文本说明和代码运行的交互方式
- 文本行和代码行不能混用
- 文本行格式是Markdown
- 每一“行”（=框）都可以任意扩充为n行内容
  - 多行情况下显示行序列号（场景：快速查错）： 菜单 > Toggle Notebook Line Numbers, 或Shift+L
- 文本行（Markdown类型）和代码行（Code类型）可以随时转换，但注意转换后的运行逻辑的改变
- 代码行的注释： # comments

### Python代码的运行
- 规则： 逐行运行。跳跃性运行会导致报错（可能但非必然）
- VS Code:
  - 清除内存（不再需要的运行结果）： restart （优于 Clear All Outputs）
  - 在任意行（位置）接龙， 需要运行所有前置代码行： Run All， 

## 2. 进阶 Advanced <a id="2"></a>

### 文本说明：
- 文字高亮(加"\`"):  `word`
- 在文本两旁加上“\**”，即可设置为粗体:  **Word**

### 代码辅助功能：使用魔法指令 Magic Instructions
- 计算代码执行时间: %timeit
- 导入Python代码,  %load filename.py, %run filename.py 
- 进入调试器: %debug
- 了解函数，指令说明, 快速查使用文档： ?name


### shell终端指令：
