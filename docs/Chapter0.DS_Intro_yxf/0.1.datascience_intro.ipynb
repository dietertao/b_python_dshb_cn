## 核心概念：
- 数据：
  - 你是哪类数据生产者（或数据关联者）？
  - 你认为的数据就是专家眼中的数据吗？
  - DataScience语义中的数据定义是什么？
    - DS视角的数据最基础的定义=未经处理的原始记录。属性：缺乏有效组织，并没有被加工成某种序列信息
    - 我们的延申理解：一切可以被我们提取价值的原始的+被处理过的符号记录。 关注数据价值属性

## 核心概念：
- 数据思维：
  - 一种以数据为中心的**思维方式**和行为范式，是用数据贯彻决策、问题解决和创新等所有过程的系统化的思考和行动方式。

- 数据/数据思维的意义：
    - **（DIKW Pyramid）模型：数据→信息→知识→智慧的转化链，实现从底层数据到顶层价值的能力跃迁**




## 数据科学的练习方法：
- 战略层：= 心法
  - 培养数据敏感性，习惯观察提高感知：一切可量化，一切可分析
  - 从数据中提炼数据价值 （阿里巴巴口号：“一切业务数据化，一切数据业务化”）
- 战役层：= 框架流程+方法论
- 战术层：= Python编程技巧

- 学习方法：
    - 重视理论，修心（培养数据思维）为上，学会慢思考 
    - 勤于查阅 （不可能一个人记住所有指令，要善于和勤于查阅，这不是low）
    - 身体力行 （代码一定要自己敲）
    - 联系实际 （一切为了实战）

## 数据科学的学习途径：
1. 阶段1：熟练使用工具
    - 阶段目标：学习Python工具，可独立完成中等规模数据的分析全流程
  
2. 阶段2：工程思维筑基
    - 阶段目标：理解领域，环境数据处理规范 = 搭建自己的数据处理流程，固化数据分析全流程

1. 阶段3：领域计算融合（持续迭代）
    - 阶段目标：成为领域里的数据分析高手

## 我们的数据思维知识的提示总结

### 数据认知的底层架构
- 数据的三重存在形态
    - 原子态：物理世界的电磁信号[0,1]（内存电磁状态）
    - 符号态：人类可解读的编码体系（ASCII/UTF-8字符集，float）
    - 语义态：有上下文含义的信息单元（温度20→舒适，股价波动10%→市场情绪）
