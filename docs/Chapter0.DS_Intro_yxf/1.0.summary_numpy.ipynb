# 学习笔记_Numpy总结



## NumPy极简框架

### 核心理念

**`数据结构`** + **`操作`**

### 核心数据结构：

ndarray（N维数组）

数学术语：
- 1维数组 = 向量 Vectors
- 2为数组 = 矩阵 Matrices
- 多维数组 = 张量 Tensors

**3大核心特征**（工厂质检标准）
1.	同质性：所有元素必须同类型 → 积木材质统一（全塑料/全金属）
np.array([1, 2.5]) → 自动升级为float类型  
2.	维度固定：创建时确定形状 → 模具不可变形（如2×3格子必须填满6个元素）
3.	连续内存：元素物理存储连续 → 流水线高效生产（对比列表的分散存储）


**快速自检方法**: 
- 维度结构 → np.array.shape
- 元素类型 → np.array.dtype  int32/int64
- 维度数 → np.array.ndim
- 元素总数 → np.array.size


## 数据操作


### 四大操作原则（流水线作业手册）

**原则1**：向量化优先（避免循环）

[反面案例]：Python循环计算平方  
squares = [x**2 for x in range(1000)]  

[NumPy正确姿势]：  
arr = np.arange(1000)  
squares = arr ** 2  # 速度提升50倍以上  

底层操作：元素基别上的操作 element-wise-operations

**原则2**：广播机制（自动扩展维度）
- 规则：从后往前对齐维度，缺失维度视为1，进行扩展


**原则3**：视图而非复制（节省内存）
- view = arr[:2, :]    # 切片创建视图（共享数据）  
- copy = arr.copy()    # 显式复制数据（内存翻倍）  

**原则4**：轴（axis）方向操作
- 口诀：
  - axis=0 → 跨行（垂直方向）
  - axis=1 → 跨列（水平方向）
  - arr.sum(axis=0)  # 每列求和 → 结果形状(3,)
