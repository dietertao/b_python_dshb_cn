#### 1.2.1. Pandas读取Text数据方法


读取以 txt格式保存的表格数据可以使用：  
pd.read_csv() - Options

>**pandas.`read_csv`**(<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;file_path_or_buffer,<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;sep='\t',<br>

参数说明：  
- file_path_or_buf: 字符串（用单/双引号），被导入的带路径的文件名，也可以是URL链接
- sep: 字符串，指定txt中的分割符. tab: '\t'
- header：整数，指定源文件作为列名的行数。默认值：0. 无列名：None
- encoding：字符串，指定文件的编码格式。默认值：None

- encoding：字符串，指定文件的编码格式。
