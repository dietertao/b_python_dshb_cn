```python
# 可以用一个数组创建Series对象

data = pd.Series([0.25, 0.5, 0.75, 1.0])
data
```

```python
# 你会发现 Series 对象将一组数据和一组索引绑定在一起，我们可以通过 values 属性和 index 属性获取数据

# data.values
# data.index

print(data.values, data.index, sep='\n') # 通过 sep 参数指定分隔符
```

```python
# 和 NumPy 数组一样，数据可以通过 Python 的中括号索引标签获取
# Pandas 的 Series 对象比它模仿的一维 NumPy 数组更加通用、灵活。

data[1]
data[1:3]
```

```python
# 索引不再仅仅是整数，还可以是任意想要的类型

data = pd.Series([0.25, 0.5, 0.75, 1.0],
                 index=['a', 'b', 'c', 'd'])

print(data)
print(f"\nc = {data['c']}")
```

```python
# 也可以使用不连续或不按顺序的索引：

data = pd.Series([0.25, 0.5, 0.75, 1.0],
                  index=[2, 5, 3, 7])

print(data)
print(f"\nNr.5 = {data[5]}")
```

```python
# 直接用 Python 的字典创建一个 Series 对象，让 Series 对象与字典的类比更加清晰

population_dict = {'California': 38332521,
                    'Texas': 26448193,
                    'New York': 19651127,
                    'Florida': 19552860,
                    'Illinois': 12882135}

# 通过字典创建 Series 对象, key 作为索引 index，value 作为数据 values
population = pd.Series(population_dict) 

population
# print(f"\nCalifornia:\t{population['California']}")
```

```python
# 和字典不同，Series 对象还支持数组形式的操作，比如切片：

population
population['California':'New York']
```
